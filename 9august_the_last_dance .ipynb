```python
import os
import pandas as pd
import matplotlib.pyplot as plt

pkl_path = os.path.join("/home/aoschu", "thesis code", "ESN_MIMO_v1", "server", "BER_results_sionna_blockfading.pkl")

# Load pickle
data = pd.read_pickle(pkl_path)

ebn0 = data['EBN0']
ber_uncoded = data['BER_uncoded']
ber_decoded = data['BER_decoded']

plt.figure(figsize=(8,6))
for key, values in ber_uncoded.items():
    plt.semilogy(ebn0, values, marker='o', label=f"{key} (Uncoded)")
for key, values in ber_decoded.items():
    plt.semilogy(ebn0, values, marker='s', linestyle='--', label=f"{key} (Decoded)")

plt.xlabel("SNR (EB/N0) [dB]")
plt.ylabel("BER")
plt.title("BER vs SNR for Different Detectors")
plt.grid(True, which='both')
plt.legend()
plt.tight_layout()
plt.show()
```
